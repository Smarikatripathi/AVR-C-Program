Exploring features of NLTK


```python
#Open the text file :
text_file = open("/lab2.txt")

#Read the data :
text = text_file.read()

#Datatype of the data read :
print (type(text))
print("\n")

#Print the text :
print(text)
print("\n")
#Length of the text :
print (len(text))
```

Sentence tokenization

```python
#Import required libraries :
import nltk
from nltk import sent_tokenize
from nltk import word_tokenize
nltk.download ('punkt')
nltk.download('punkt_tab')
```

```python
#Tokenize the text by sentences :
sentences = sent_tokenize(text)

#How many sentences are there? :
#print (len(sentences))

#Print the sentences :
print(sentences)
sentences
```

Word tokenizing

```python
#Tokenize the text with words :
words = word_tokenize(text)

#How many words are there? :
print (len(words))

#Print words :
print (words)
```

Find the frequency distribution:

```python
#Import required libraries :
from nltk.probability import FreqDist

#Find the frequency :
fdist = FreqDist(words)

#Print 10 most common words :
fdist.most_common(10)
```

Plot the frequency graph:

```python
#Plot the graph for fdist :
import matplotlib.pyplot as plt

fdist.plot(10)
```

Remove punctuation marks:

```python
#Empty list to store words:
words_no_punc = []

#Removing punctuation marks :
for w in words:
    if w.isalpha():
        words_no_punc.append(w.lower())

#Print the words without punctution marks :
print (words_no_punc)

print ("\n")

#Length :
print (len(words_no_punc))
```

Displaying words without punctuation

```python
#Frequency distribution :
fdist = FreqDist(words_no_punc)

fdist.most_common(10)
```

Plotting graph without punctuation marks:

```python
fdist.plot(10)
```

Importing stopwords from NLTK

```python
nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords

#List of stopwords
stopwords = stopwords.words("english")
print(stopwords)
```

Removing stopwords:

```python
#Empty list to store clean words :
clean_words = []

for w in words_no_punc:
    if w not in stopwords:
       clean_words.append(w)

print(clean_words)
print("\n")
print(len(clean_words))
```

Final frequency distribution:

```python
#Frequency distribution :
fdist = FreqDist(clean_words)

fdist.most_common(10)
```

```python
fdist.plot(10)
```

Word Cloud

```python
#Library to form wordcloud :
from wordcloud import WordCloud

#Library to plot the wordcloud :
import matplotlib.pyplot as plt

#Generating the wordcloud :
wordcloud = WordCloud().generate(text)

#Plot the wordcloud :
plt.figure(figsize = (12, 12))
plt.imshow(wordcloud)

#To remove the axis value :
plt.axis("off")
plt.show()
```

Stemming

```python
#Stemming Example :

#Import stemming Library :
from nltk.stem import PorterStemmer

porter = PorterStemmer()

#Word-list for stemming :
word_list = ["Study","Studying","Studies","Studied"]

for w in word_list:
    print(porter.stem(w))
```

```python
#Stemming Example :

# Import the stemming library
from nltk.stem import PorterStemmer

# Create the stemmer object
porter = PorterStemmer()

# Word-list for stemming
word_list = ["studies", "leaves", "decreases", "plays"]

for w in word_list:
    print(w, "", porter.stem(w))
```
